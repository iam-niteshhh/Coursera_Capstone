# 1.Data Gathering 

```python
toronto_data=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',).text
```

```python
toronto_data=BeautifulSoup(toronto_data,'html.parser')
print(toronto_data.prettify())
```

```python
table_contents=[]
table=toronto_data.find('table')
table
```

```python
for row in table.find_all('td'):
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        cell['Postal Code'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)

print(table_contents)
```

```python
data=pd.DataFrame(table_contents)
data
```

```python
data.Borough.unique()
```

```python
data['Borough']=data['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                         'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                         'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                         'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
data.Borough.unique()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            Client_ID, 
            Client_Secret, 
            Version, 
            lat, 
            lng, 
            radius, 
            100)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
downtown_venues=getNearbyVenues(downtown_data['Neighborhood'],
                      downtown_data['Latitude'],
                      downtown_data['Longitude'])
```

```python
downtown_venues
```

```python
downtown_venues_onehot=pd.get_dummies(downtown_venues['Venue Category'],prefix="",prefix_sep="")
downtown_venues_onehot['Neighbourhood']=downtown_venues['Neighborhood']

fixed=[downtown_venues_onehot.columns[-1]]+list(downtown_venues_onehot.columns[:-1])
downtown_venues_onehot=downtown_venues_onehot[fixed]
downtown_venues_onehot.head()
```

```python
downtown_grouped = downtown_venues_onehot.groupby('Neighbourhood').mean().reset_index()
downtown_grouped.head()
```

```python
num_top_venues=5
for hood in downtown_grouped['Neighbourhood']:
    print("****"+hood+"****")
    temp = downtown_grouped[downtown_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))

print(temp)
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))


neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighbourhood'] = downtown_grouped['Neighbourhood']

for ind in np.arange(downtown_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(downtown_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

```python
kcluster=5
downtown_grouped_clustering = downtown_grouped.drop('Neighbourhood', 1)

kmeans=KMeans(n_clusters=kcluster,init='k-means++',random_state=0)
```

```python
kmeans.fit(downtown_grouped_clustering)
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
downtown_merged = downtown_data
downtown_merged = downtown_merged.join(neighborhoods_venues_sorted.set_index('Neighbourhood'), on='Neighborhood')
downtown_merged.head()
```

```python
map_cluster=folium.Map(location=[g.latitude,g.longitude],zoom_start=13)

x=np.arange(kcluster)
color_array=cm.rainbow(np.linspace(0,1,len(x)))
rainbow=[colors.rgb2hex(i) for i in color_array]

for i,X in downtown_merged.iterrows():
    label=folium.Popup(str(X['Neighborhood'])+'Cluster'+str(X['Cluster Labels']),parse_html=True)
    folium.CircleMarker(location=[X['Latitude'],X['Longitude']],
                       radius=5,
                       color=rainbow[int(X['Cluster Labels'])-1],
                       fill=True,
                       fill_color=rainbow[int(X['Cluster Labels'])-1],
                       popup=label,
                       ).add_to(map_cluster)
map_cluster
    
```
